```python
for column in df.columns:
    print(f"--- Value Counts for column: {column} ---")
    print(df[column].value_counts())
    print() # Adds a blank line for readability
```

```python
# 2. Identify columns with only one unique value
# We use a list comprehension to build a list of columns to drop.
cols_to_drop = [col for col in df.columns if df[col].nunique() == 1]

len(cols_to_drop)
```

```python
print(f"\nColumns to drop: {cols_to_drop}")
```

```python

# 3. Drop the identified columns from the DataFrame
df_cleaned = df.drop(columns=cols_to_drop)

print("\nCleaned DataFrame:")
print(df_cleaned)
```

```python
df_cleaned=df_cleaned.drop('AE_172',axis=1)
```

```python
df_cleaned=df_cleaned.iloc[:,1:]
```
