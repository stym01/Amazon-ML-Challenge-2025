```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('compressed_features_vit_test.csv')
```

```python
df['Value'].isnull().sum()
```

```python
# 2. Get 'sample_id's where 'Value' is null and store in an array
removed_ids = df[df['Value'].isnull()]['sample_id'].to_numpy()

removed_ids
```

```python

# 3. Drop the rows with null values
df_cleaned = df.dropna()

# 1. Convert the NumPy array to a pandas DataFrame
removed_df = pd.DataFrame(removed_ids, columns=['sample_id'])

# 2. Save the new DataFrame to a CSV file
# We use index=False to avoid writing the DataFrame index (0, 1, 2...) as a column
removed_df.to_csv('removed_ids.csv', index=False)
```
