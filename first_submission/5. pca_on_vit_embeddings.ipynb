```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Load dataset
df = pd.read_csv("vit_test_features.csv")
old_df=df
```

```python
df.head()
```

```python

# Drop unwanted columns
# drop_cols = ['Value', 'items_link', 'sample_id', 'Unit_grams', 'Unit_ml']
# X = df.drop(columns=drop_cols + ['price'], errors='ignore')  # drop price too from features

X=df.drop(columns='sample_id')
X

```

```python
# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply PCA to select top 300 components
n_components = 300
pca = PCA(n_components=n_components)
X_pca = pca.fit_transform(X_scaled)

```

```python
# Create DataFrame with PCA features
pca_columns = [f'PC_{i+1}' for i in range(n_components)]
pca_df = pd.DataFrame(X_pca, columns=pca_columns)

pca_df
```

```python
# Add price column at the end
pca_df['sample_id'] = old_df['sample_id'].values
```

```python

# Save to CSV
pca_df.to_csv('pca_300_test_vit_features.csv', index=False)

print("CSV saved: pca_300_features.csv")
print(f"Explained variance by top 300 components: {sum(pca.explained_variance_ratio_):.4f}")
print(pca_df.head())
```
