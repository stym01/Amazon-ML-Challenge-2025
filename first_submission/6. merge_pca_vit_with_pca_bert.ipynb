```python
import pandas as pd
df_vit = pd.read_csv("pca_300_test_vit_features.csv")
df_vit.tail()
```

```python

# Remove duplicate rows based on 'sample_id'
df_vit = df_vit.drop_duplicates(subset='sample_id', keep='first')

df_bert = pd.read_csv("pca_300_test_bert_features.csv")
```

```python

merged_df = pd.merge(df_bert, df_vit, on='sample_id', how='inner')
```

```python
sample=merged_df['sample_id']
```

```python
merged=merged_df.drop(columns='sample_id')
```

```python
merged['sample_id']=sample
merged
```
